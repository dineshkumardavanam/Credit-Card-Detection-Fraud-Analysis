```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv("/content/creditcard.csv")
```

```python
credit_card_data.head()
```

```python
credit_card_data.info()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
legit.head()
```

```python
legit['Class'].value_counts()
```

```python
fraud.head()
```

```python
fraud['Class'].value_counts()
```

```python
legit_sample = legit.sample(n=492)
```

```python
legit_sample.head()
```

```python
fraud.shape
```

```python
legit_sample.shape
```

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0) #concat rowwise
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

#Accuracy on Training Data

```python
training_accuracy_score = accuracy_score(X_train_prediction,Y_train)
```

```python
training_accuracy_score
```

```python
print("The accuracy score is:",int(training_accuracy_score*100),"%")
```

#Accuracy on Test Data

```python
X_test_prediction=model.predict(X_test)
```

```python
testing_accuracy_score=accuracy_score(X_test_prediction,Y_test)
```

```python
print("The accuracy score is:",int(testing_accuracy_score*100),"%")
```
